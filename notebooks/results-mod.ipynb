```python
%matplotlib widget
import os
import pickle
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from IPython.display import Image
from tensorflow.python.framework import ops
from tensorflow.examples.tutorials.mnist import input_data
from clustereg import NeuralNetwork
network_names = ["network_scatter_embedding", "network_vanilla", "network_l2", "network_scatter_weights"]
mnist = input_data.read_data_sets("MNIST_data", one_hot=True)
mnist_fashion = input_data.read_data_sets("MNIST_data_fashion", one_hot=True)
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
```

```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = "0"
```

# Executing below cell will re-train all 4 networks for mnist (optional)

```python
nets = []
# mnist = input_data.read_data_sets("MNIST_data", one_hot=True)
# ------------------------------------------------------------                           
tf.reset_default_graph()                                                                                                                                           
ops.reset_default_graph()
for net in network_names:
    nn = NeuralNetwork(mnist, net, epochs=126, batch_size=64, dataset="mnist")
    epochs, network = nn.train()
    nets.append(network)
#     Accuracy and Loss metrics
    nn.plot_metrics(epochs=epochs, nets=nets, metric='train', title="Train Acc. / Loss")
    nn.plot_metrics(epochs=epochs, nets=nets, metric='valid', title="Valid. Acc. / Loss")
    nn.plot_metrics(epochs=epochs, nets=nets, metric='test', title="Test Acc. / Loss")
#     Calibration
    val_logits, val_labels = nn.predict(epoch=0, mode='valid', verbose=False)
    nn.reliability_diagram(logits=val_logits, labels=val_labels)
    val_logits_calib, val_labels_calib = nn.set_temperature(logits=val_logits, labels=val_labels)
    nn.reliability_diagram(logits=val_logits_calib, labels=val_labels_calib, scaled=True)                    
    nn.sess.close()                                                                                                                                                    
    tf.reset_default_graph()                                                                                                                                           
    ops.reset_default_graph()
    plt.close('all')
    del nn.sess, nn
with open("./logs/mnist_nets_logs.pickle", "wb") as fh:
    pickle.dump(nets, fh)
# del mnist
del nets
```

## Load checkpoint with lowest validation loss

```python
d = {}
for net in network_names:
    d[net] = os.listdir(os.path.join(os.getcwd(), f"logs/checkpoints/mnist/{net}"))
df = pd.concat([pd.DataFrame([d[net] for net in network_names], index=[n for n in d])], axis=1).T
display(df)
```

```python
best_mnist_checkpoints = []
best_mnist_checkpoints.append(df.iloc[67, 0])
best_mnist_checkpoints.append(df.iloc[44, 1])
best_mnist_checkpoints.append(df.iloc[40, 2])
best_mnist_checkpoints.append(df.iloc[5, 3])
epochs = [int(val.split("_")[1]) for val in best_mnist_checkpoints]
```

# Load checkpointed models & recalculate stats for valid, test on mnist

```python
valid_acc = []
valid_loss =[]
test_acc = []
test_loss = []
tf.reset_default_graph()
ops.reset_default_graph()
for idx, net in enumerate(network_names):
    sess = tf.Session()
    sess.run(tf.global_variables_initializer())
    saver = tf.train.import_meta_graph(f"./logs/checkpoints/mnist/{net}/{best_mnist_checkpoints[idx]}", clear_devices=True)
    saver.restore(sess, f"./logs/checkpoints/mnist/{net}/" + '.'.join(best_mnist_checkpoints[idx].split(".")[:-1]))
    graph = tf.get_default_graph()
    x = graph.get_tensor_by_name("data:0")
    y = graph.get_tensor_by_name("labels:0")
#     logits = graph.get_tensor_by_name(f"{net}/layer-3/identity-l3:0")
#     lgts = sess.run(logits, feed_dict={x: mnist.validation.images})
#     import pdb; pdb.set_trace()
#     probabilities = graph.get_tensor_by_name(f"{net}/layer-3/softmax_probabilities:0")
#     final_loss = graph.get_tensor_by_name(f"{net}/final_loss:0")
    accuracy = graph.get_tensor_by_name(f"{net}/metrics/accuracy/accuracy:0")
    val_acc = sess.run(accuracy, feed_dict={x: mnist.validation.images, y: mnist.validation.labels})
    valid_acc.append(val_acc)
    te_acc = sess.run(accuracy, feed_dict={x: mnist.test.images, y: mnist.test.labels})
    test_acc.append(te_acc)
    sess.close()
    tf.reset_default_graph()
    ops.reset_default_graph()
    del sess
```

# Accuracy vs. Loss for (train, valid, test) --> for comparing test + valid stats with above cell

```python
with open("./logs/mnist_nets_logs.pickle", "rb") as fh:
    mnist_stats = pickle.load(fh)

tmp_network_names = []
train_acc = []
train_loss = []
valid_acc = []
valid_loss = []
test_acc = []
test_loss = []
negloglike_before_calibration = []
ece_before_calibration = []
negloglike_after_calibration = []
ece_after_calibration = []

for net_stat, epoch in zip(mnist_stats, epochs):
    tmp_network_names.append(net_stat['title']) 
    train_acc.append(net_stat['train']['acc'][epoch])
    train_loss.append(net_stat['train']['loss'][epoch])
    valid_acc.append(net_stat['valid']['acc'][epoch])
    valid_loss.append(net_stat['valid']['loss'][epoch])
    test_acc.append(net_stat['test']['acc'][int(epoch / 25) - 1])
    test_loss.append(net_stat['test']['loss'][int(epoch / 25) - 1])
    negloglike_before_calibration.append(net_stat['calibration']['before']['NLL'])
    ece_before_calibration.append(net_stat['calibration']['before']['ECE'])
    negloglike_after_calibration.append(net_stat['calibration']['after']['NLL'])
    ece_after_calibration.append(net_stat['calibration']['after']['ECE'])

d = {('Accuracy','train'): train_acc, ('Accuracy','valid'): valid_acc, ('Accuracy','test'): test_acc,
     ('Loss','train'): train_loss, ('Loss','valid'): valid_loss, ('Loss','test'): test_loss}
df = pd.DataFrame(d, index=tmp_network_names)
# df.columns.names = ('l1','l2')
del tmp_network_names
df
```

# Scatter plots of penultimate layer for 1K training samples

```python
tf.reset_default_graph()
ops.reset_default_graph()
for idx, net in enumerate(network_names):
    sess = tf.Session()
    sess.run(tf.global_variables_initializer())
    saver = tf.train.import_meta_graph(f"./logs/checkpoints/mnist/{net}/{best_mnist_checkpoints[idx]}", clear_devices=True)
    saver.restore(sess, f"./logs/checkpoints/mnist/{net}/" + '.'.join(best_mnist_checkpoints[idx].split(".")[:-1]))
    graph = tf.get_default_graph()
    x = graph.get_tensor_by_name("data:0")
    penultimate = graph.get_tensor_by_name(f"{net}/layer-2/relu-l2:0")
    logits = graph.get_tensor_by_name(f"{net}/layer-3/identity-l3:0")
    probabilities = graph.get_tensor_by_name(f"{net}/layer-3/softmax_probabilities:0")
    lgts, probs, penult = sess.run([logits, probabilities, penultimate], feed_dict={x: mnist.train.images[100:1100, :]})
    sess.close()
    tf.reset_default_graph()
    ops.reset_default_graph()
    del sess
    nn = NeuralNetwork(mnist, net, epochs=126, batch_size=1000, dataset="mnist")
    nn.plot_scatter(penult, mnist.train.labels[100:1100].argmax(axis=1), probs.max(axis=1), epoch=0.000)
    nn.sess.close()
    tf.reset_default_graph()
    ops.reset_default_graph()
    del nn.sess, nn
    
    with open("{}/mnist_1.pickle".format(net), "rb") as fp:
        fig = pickle.load(fp)
        fig.show()
```
